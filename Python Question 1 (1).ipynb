```python
class BudgetCalculator:
    def __init__(self, user_code, gross_income):
    # Initialize the BudgetCalculator with the user code and gross income
     self.user_code = user_code
     self.gross_income = gross_income
     self.tax = self.calculate_tax()
     self.income_after_tax = self.gross_income - self.tax
     self.expenses = self.calculate_expenses()
     self.total_expenses = sum(self.expenses.values())
     self.net_income = self.income_after_tax - self.total_expenses

    def calculate_tax(self):
        #Calculate the tax based on the tax brackets
        income = self.gross_income
        if income <= 237100:
            tax = income * 0.18
        elif income <= 370500:
            tax = 42678 + (income - 237100) * 0.26
        elif income <= 512800:
            tax = 77362 + (income -370500) * 0.31
        elif income <=67300:
            tax = 121475 + ( income - 512800) *0.36
        elif income <= 857900:
            tax = 179147 + (income - 673000) * 0.39
        elif income <= 1817000:
            tax = 251258 + (income - 857900) *0.41
        else:
            tax = 644489 + (income - 1817000) *0.45
        return tax

    def calculate_expenses(self):
    #calculate the expenses for each category based on the percentages
        expenses = {
            'Utilities': self.income_after_tax * 0.05,
            'Rent': self.income_after_tax * 0.15,
            'Transport': self.income_after_tax * 0.30,
            'Health': self.income_after_tax * 0.003,
            'Groceries': self.income_after_tax * 0.10,
            'Communication': self.income_after_tax * 0.02,
        }
        return expenses

    def display_budget(self):
    #Display the monthly budget with all calculated values
        print("\nMonthly Budget for User:", self.user_code)
        print("  ")
        print("Income")
        print("Gross Monthly Income (Before Tax): R", self.gross_income)
        print("Gross Monthly Incone (After Tax): R", self.income_after_tax)
        print("\n  ")
        print("\nExpenses:")
        print("  ")
        for category, amount in self.expenses.items():
            print(f"Total Expenses: R{amount:.2f}")
        print(f"Total Expenses: R{self.total_expenses:.2f}")
        print("\n  ")
        print("Net Income: R", self.net_income)

def main_menu():
        while True:
            print("\nBudget Portal")
            print("1. Create new entry")
            print("0. Exit")
            option = input("Select an option: ")
            if option == '1':
                #prompt the user for user code and grross income
                user_code = input("Enter user code: ")
                gross_income = float(input("Enter gross income before tax: "))
                budget = BudgetCalculator(user_code, gross_income)
                budget.display_budget()
            elif option == '0':
                print("Exiting the program")
                break
            else:
                #Handle invalid options
                print("Invalid option. Please try again.")

if __name__ == "__main__":
    #Run the main menu when the script is executed
    main_menu()
        
        
        
```
